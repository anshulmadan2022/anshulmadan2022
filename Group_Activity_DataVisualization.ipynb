1. Basic Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

def basic_line_plot():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title('Basic Line Plot: Sine Wave')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.grid(True)
    plt.show()
basic_line_plot()
```

3. Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plot():
    x = np.random.rand(50)
    y = np.random.rand(50)
    colors = np.random.rand(50)
    sizes = 1000 * np.random.rand(50)
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
    plt.title('Scatter Plot with Color and Size Variation')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.colorbar()
    plt.show()
scatter_plot()
```

4. Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

def bar_plot():
    categories = ['A', 'B', 'C', 'D', 'E']
    values = [23, 35, 17, 29, 12]
    plt.figure(figsize=(10, 6))
    plt.bar(categories, values)
    plt.title('Bar Plot')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    for i, v in enumerate(values):
        plt.text(i, v, str(v), ha='center', va='bottom')
    plt.show()
bar_plot()
```

5. Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

def histogram():
    data = np.random.randn(1000)
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, edgecolor='black')
    plt.title('Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
histogram()
```

6. Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

def subplots():
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle('Subplots Example')

    # Subplot 1: Line plot
    x = np.linspace(0, 10, 100)
    axs[0, 0].plot(x, np.sin(x))
    axs[0, 0].set_title('Sine Wave')

    # Subplot 2: Scatter plot
    x = np.random.rand(50)
    y = np.random.rand(50)
    axs[0, 1].scatter(x, y)
    axs[0, 1].set_title('Scatter Plot')

    # Subplot 3: Bar plot
    categories = ['A', 'B', 'C', 'D']
    values = [23, 35, 17, 29]
    axs[1, 0].bar(categories, values)
    axs[1, 0].set_title('Bar Plot')

    # Subplot 4: Histogram
    data = np.random.randn(1000)
    axs[1, 1].hist(data, bins=30, edgecolor='black')
    axs[1, 1].set_title('Histogram')

    plt.tight_layout()
    plt.show()
subplots()
```

7. Customised Plots

```python
import matplotlib.pyplot as plt
import numpy as np

def customized_plot():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)

    plt.figure(figsize=(12, 7))
    plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, markevery=10)
    plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2, marker='s', markersize=6, markevery=10)

    plt.title('Customized Plot: Sine and Cosine', fontsize=16, fontweight='bold')
    plt.xlabel('X axis', fontsize=12)
    plt.ylabel('Y axis', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True, linestyle=':', alpha=0.7)

    plt.xlim(0, 10)
    plt.ylim(-1.5, 1.5)

    plt.axhline(y=0, color='k', linestyle='--', linewidth=1)
    plt.axvline(x=5, color='k', linestyle='--', linewidth=1)

    plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.3),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    plt.show()
customized_plot()
```

8. Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap

def pie_chart():
    labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
    sizes = [15, 30, 45, 10]
    explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice

    fig, ax = plt.subplots(figsize=(10, 7))
    ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
           shadow=True, startangle=90)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.title("Pie Chart")
    plt.show()
pie_chart()
```

9. Stacked Area Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap

def stacked_area_plot():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.tan(x)

    plt.figure(figsize=(10, 6))
    plt.stackplot(x, y1, y2, y3, labels=['sin(x)', 'cos(x)', 'tan(x)'])
    plt.title('Stacked Area Plot')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()
stacked_area_plot()
```

10. 3D Surface Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap

def surface_plot():
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    x = np.arange(-5, 5, 0.25)
    y = np.arange(-5, 5, 0.25)
    x, y = np.meshgrid(x, y)
    r = np.sqrt(x**2 + y**2)
    z = np.sin(r)

    surf = ax.plot_surface(x, y, z, cmap='viridis')
    ax.set_title('3D Surface Plot')
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
surface_plot()
```
